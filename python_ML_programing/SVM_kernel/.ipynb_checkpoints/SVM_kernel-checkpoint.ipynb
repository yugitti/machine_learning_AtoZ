```python
from sklearn import  datasets
plt.show()
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'rbf', C=1.0, gamma=0.2,random_state=1)
svm.fit(X_train_std, y_train)
%run myGraph.py
plot_decision_regions(X=X_combined_std, y=y_combined, classifier=svm, test_idx=range(105,150))
plt.tight_layout()
```

### C

```python
weights, params, predict = [], [], []
for c in np.arange(-5,5):
    c = 10.**c
    svm = SVC(kernel = 'rbf', C=c, gamma=0.2,random_state=1)
    svm.fit(X_train_std, y_train)
    params.append(c)
    
    y_pred = svm.predict(X_test_std)
    y_c = pd.DataFrame(np.c_[y_pred, y_test], columns={'predict', 'original'})
    predict.append(round((y_c['predict'] == y_c['original']).sum() / len(y_c['predict']) * 100,0))
```

```python
fig = plt.figure(figsize=(10,5),dpi=100)
ax = plt.axes()
ax.plot(params, predict, label = 'accuracy')
ax.set_ylabel('accuracy')
ax.set_xlabel('C')
ax.legend(loc='upper left')
ax.set_xscale('log')
```
